```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler,LabelEncoder
from sklearn.pipeline import Pipeline
from sklearn.decomposition import PCA
```

```python
main_data = pd.read_csv("breast-cancer.csv")
```

```python
df = main_data.drop('id',axis=1)
```

```python
x = df.drop(['diagnosis'],axis=1)
y = df["diagnosis"]
x_train , x_test , y_train , y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
x_train
```

```python
x_train.describe()
```

```python
trf = [
    ('trf1',StandardScaler()),
    ('trf2',PCA(n_components=10))
]
```

```python
pipe = Pipeline(steps=trf)
```

```python
x_train_transformed = pd.DataFrame(pipe.fit_transform(x_train))
x_test_transformed = pd.DataFrame(pipe.transform(x_test))
```

```python
le = LabelEncoder()
```

```python
y_train_transformed = pd.DataFrame(le.fit_transform(y_train))
y_test_transformed = pd.DataFrame(le.transform(y_test))
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis, QuadraticDiscriminantAnalysis
from sklearn.neural_network import MLPClassifier
from sklearn.gaussian_process import GaussianProcessClassifier
from sklearn.gaussian_process.kernels import RBF
from sklearn.calibration import CalibratedClassifierCV
from sklearn.linear_model import SGDClassifier
from sklearn.ensemble import AdaBoostClassifier
from sklearn.ensemble import BaggingClassifier
from sklearn.metrics import precision_score,recall_score,f1_score
```

```python
models = {
    'LogisticRegression': LogisticRegression(),
    'KNN': KNeighborsClassifier(),
    'DecisionTree': DecisionTreeClassifier(),
    'RandomForest': RandomForestClassifier(),
    'GradientBoosting': GradientBoostingClassifier(),
    'SVC': SVC(),
    'GaussianNB': GaussianNB(),
    'LDA': LinearDiscriminantAnalysis(),
    'QDA': QuadraticDiscriminantAnalysis(),
    'MLPClassifier': MLPClassifier(max_iter=1000),
    'GaussianProcess': GaussianProcessClassifier(1.0 * RBF(1.0)),
    'CalibratedSVC': CalibratedClassifierCV(SVC(), method='isotonic'),
    'SGDClassifier': SGDClassifier(penalty='l1',alpha=0.01),
    'AdaBoostClassifier': AdaBoostClassifier(),
    'BaggingClassifier':BaggingClassifier()
}
```

```python
result=[]
```

```python
for name , model in models.items():
    model.fit(x_train_transformed,y_train_transformed)
    y_pred = model.predict(x_test_transformed)
    acc = accuracy_score(y_test_transformed,y_pred)
    prec_score = precision_score(y_test_transformed,y_pred)
    recall_Score = recall_score(y_test_transformed,y_pred)
    f1_Score = f1_score(y_test_transformed,y_pred)
    result.append((name,acc,prec_score,recall_Score,f1_Score))
```

```python
result = pd.DataFrame(result)
```

```python
new_column = {
    0: 'Model Name',
    1: 'Accuracy',
    2: 'Precision',
    3: 'Recall',
    4: 'F1-Score'
}
```
